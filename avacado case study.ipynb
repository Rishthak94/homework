# Normalization

```python
import scipy.stats as stat
import pylab
```

```python
def plot_data(df,feature):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    df[feature].hist()
    plt.subplot(1,2,2)
    stat.probplot(df[feature],dist='norm',plot=pylab)
    plt.show()
```

```python
plot_data(df,'Total Volume')
```

```python
df['Total Volume_reciprocal']=1/df['Total Volume']
plot_data(df,'Total Volume_reciprocal')
```

```python
df['Total Volume_square']=df['Total Volume']**(1/2)
plot_data(df,'Total Volume_square')
```

```python
df['Total Volume_boxcox'],parameters=stat.boxcox(df['Total Volume'])
```

```python
plot_data(df,'Total Volume_boxcox')
```
